```python
import os
folder_path = r"C:\Users\koush\Desktop\GenAi project folder\CSV"
all_files = os.listdir(folder_path)
non_genAi_files=[f for f in all_files if f.startswith("non")]
genAi_files=[f for f in all_files if f.startswith("genai")]
print("genAi files", len(genAi_files))
print("non genAi files", len(non_genAi_files))
```

```python
rows = []
```

```python
for file in genAi_files:
    f = file.replace(".csv","")
    f = f.split("_")

    label = "genai"
    app = f[1]
    mode = f[2]
    run_id = f[3]
    category= None

    rows.append({
        "file": file,
        "label": label,
        "category": category,
        "app": app,
        "mode": mode,
        "run_id": run_id
    })
    
    
```

```python
for file in non_genAi_files:
    f = file.replace(".csv","")
    f = f.split("_")

    if f[2] in ["audio","video"] and f[3]=="calling":
        label = "non_genai"
        category = "calling"
        app = f[4]
        run_id = f[5]
        mode= f[2]
    else:
        label = "non_genai"
        category = f[2]
        app = f[3]
        run_id = f[4]
        mode= None



    rows.append({
        "file": file,
        "label": label,
        "category": category,
        "app": app,
        "mode": mode,
        "run_id": run_id
    })
```
